### Analysis

- Looking at the boxplot one would assume:
    - Capomulin and Ramicane are more effective in reducing tumor size than Infubinol and Ceftamin.
    - Infubinol and Ceftamin have almost double the Final Tumor Volume that Capomulin and Ramicane have.
- But in reality:
    - Infubinol and Ceftamin have significantly higher mean tumor volume than Capomulin and Ramicane.
- The more a mouse weighs, the larger their tumor is.
    - The correlation between mouse weight and the average tumor volume is 0.84

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="right")

# Display the data table for preview
reduced_merge_df = merge_df.loc[:, ["Mouse ID", "Timepoint", "Tumor Volume (mm3)",
                                "Metastatic Sites", "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]
reduced_merge_df.head()
```

```python
# Checking the number of mice.
len(reduced_merge_df["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
unique_mice_df = reduced_merge_df.groupby(["Mouse ID", "Timepoint"])

# Count the occurrences of each value in column 'Mouse ID'
counts = unique_mice_df['Mouse ID'].value_counts()

# Filter to get the values with counts greater than 1
result = counts[counts > 1]
result
```

```python
# Optional: Get all the data for the duplicate mouse ID.
reduced_merge_df.loc[(reduced_merge_df["Mouse ID"] == "g989")]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
reduced_merge_df = reduced_merge_df[reduced_merge_df["Mouse ID"] != "g989"]
reduced_merge_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(reduced_merge_df["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
reduced_merge_df.groupby(["Drug Regimen"])

mean_tumor_volume = reduced_merge_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_tumor_volume = reduced_merge_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
variance_tumor_volume = reduced_merge_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
stddev_tumor_volume = reduced_merge_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
stderr_tumor_volume = reduced_merge_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
# Assemble the resulting series into a single summary DataFrame.
drug_regimen = pd.DataFrame({
    "Mean Tumor Volume" : mean_tumor_volume,
    "Median Tumor Volum" : median_tumor_volume,
    "Tumor Volume Variance" : variance_tumor_volume,
    "Tumor Volume Std. Dev." : stddev_tumor_volume,
    "Tumor Volume Std. Err." : stderr_tumor_volume
})
drug_regimen
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
# Create a DataFrame
summary_statistics = reduced_merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', lambda x: x.std() / (len(x) ** 0.5)]).rename(columns={'<lambda_0>': 'sem'})
summary_statistics
# Aggregate the data
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_mouse_timepoint = reduced_merge_df.groupby("Mouse ID")['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
tumor_volume_latest_timepoint = pd.merge(max_mouse_timepoint, reduced_merge_df, how="left")
tumor_volume_latest_timepoint = tumor_volume_latest_timepoint.rename(columns={"Timepoint": "Latest Timepoint"})

#Filter for these four treatment regimens: Capomulin, Ramicane, Infubinol, and Ceftamin
four_treatment_regimens = tumor_volume_latest_timepoint[tumor_volume_latest_timepoint['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Display the data table for preview
four_treatment_regimens
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data = []
lower_bounds = []
upper_bounds = []

# Loop through each treatment
for treatment in treatment_list:
    # Filter the DataFrame for the current treatment
    treatment_data = four_treatment_regimens[four_treatment_regimens['Drug Regimen'] == treatment]
    
    # Locate the rows which contain mice on each drug and get the tumor volumes and convert to Series
    # add subset
    tumor_volume = treatment_data["Tumor Volume (mm3)"].tolist()
    tumor_volume_series = pd.Series(tumor_volume)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    quartiles = tumor_volume_series.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    
    # Append results to lists
    tumor_volume_data.append(tumor_volume)
    lower_bounds.append(lower_bound)
    upper_bounds.append(upper_bound)

    # Print results for each treatment
    print(f"{treatment} - Lower Quartile: {lowerq}, Upper Quartile: {upperq}, IQR: {iqr}")
    print(f"{treatment} - Values below {lower_bound} could be outliers.")
    print(f"{treatment} - Values above {upper_bound} could be outliers.\n")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
flierprops = dict(marker='o', markerfacecolor='red', markersize=12, markeredgecolor='black')
plt.boxplot(tumor_volume_data, flierprops=flierprops)
plt.ylabel("Final Tumor Volume (mm3)")
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = reduced_merge_df[reduced_merge_df["Drug Regimen"] == "Capomulin"]
capomulin_stats = capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().reset_index()
capomulin_stats = capomulin_df.groupby("Mouse ID").agg(Average_Tumor_Volume=('Tumor Volume (mm3)', 'mean'),Weight=('Weight (g)', 'first')
                                                      ).reset_index()

x_axis = capomulin_stats["Weight"]
y_axis = capomulin_stats["Average_Tumor_Volume"]
plt.scatter(x_axis, y_axis)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
correlation = st.pearsonr(x_axis,y_axis)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

plt.scatter(x_axis, y_axis)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

slope, intercept, r_value, p_value, std_err = stats.linregress(x_axis, y_axis)
regression_line = slope * x_axis + intercept
plt.plot(x_axis, regression_line, color='red')

plt.show()
```
